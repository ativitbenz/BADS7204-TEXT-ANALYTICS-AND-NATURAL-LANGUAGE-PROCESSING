```python
file = os.getcwd()+'\\train_cut_poem.txt'
word_list = []
with open(file, 'r', encoding="utf-8") as f:
    for line in f.read().split('\n'):
        word_list.extend(line.split('|'))
len(word_list)
```

```python
k = list(map(len, word_list))
sum(k)/len(k)
```

```python
d = dict(zip(word_list,k))
```
